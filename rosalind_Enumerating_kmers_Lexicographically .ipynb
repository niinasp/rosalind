Organizing Strings

When cataloguing a collection of genetic strings, we should have an established system by which to organize them.
The standard method is to organize strings as they would appear in a dictionary, so that "APPLE" precedes "APRON", 
which in turn comes before "ARMOR".

Problem

Assume that an alphabet A has a predetermined order; that is, we write the alphabet as a permutation A=(a1,a2,…,ak), 
where a1<a2<⋯<ak. For instance, the English alphabet is organized as (A,B,…,Z).

Given two strings s and t having the same length n, we say that s precedes t in the lexicographic order (and write s<Lext)
if the first symbol s[j] that doesn't match t[j] satisfies sj<tj in A.

Given: A collection of at most 10 symbols defining an ordered alphabet, and a positive integer n
(n≤10).

Return: All strings of length n that can be formed from the alphabet, ordered lexicographically 
(use the standard order of symbols in the English alphabet).

```python
import os
from itertools import product

# give path and filename
path = "C:/Users/Niina/Google Drive/mdp_bioinformatics/rosalind/bioinformatics stronghold/"
os.chdir(path)
f = open("lex.txt")

# read letters and number into a list
r = []
for line in f:
    line = line.rstrip()
    r.append(line)
f.close()

k = int(r[1]) # collection of symbols
a = r[0].split()  # integer

# make list that includes k times the list of characters
strings1 = []
for k in range(0, k):
    strings1.append(a)
    
# collect all combinations into another list and sort it
strings2 = []
for items in product(*strings1):
    strings2.append("".join(items))
strings2.sort()    

# print combinations
for c in strings2:
    print(c)
```
